```python
import ccxt
import pandas as pd
getattr(ccxt,'binance')()
binance = ccxt.binance()
```

```python
def getprices(exchange, symbol):
    inst = getattr(ccxt, exchange)()
    df = pd.DataFrame(inst.fetchOHLCV(symbol, '1h', limit = 1000000))
    df.columns = ['Time', 'Open', 'High','Low','Close', 'Volumne']
    df.set_index('Time',inplace=True)
    df.index = pd.to_datetime(df.index,unit = 'ms')
    df = df.astype(float)
    return df
getprices('binance','BTC/USDT')
```
